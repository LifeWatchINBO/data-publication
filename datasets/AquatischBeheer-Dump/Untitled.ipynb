#**Aquatisch Beheer Conversie Coordinaten**

```python
import numpy as np
import pandas as pd
```

```python
survey_data = pd.read_csv("VermeerschCompleetOccIDRefine.csv")
```

```python
survey_data.head()
```

```python
import pyproj
```

```python
lambert = pyproj.Proj("+init=EPSG:31370")
wgs84 = pyproj.Proj("+init=EPSG:4326")
```

```python
pyproj.transform(lambert, wgs84,221651,44263)
```

```python
def transform_lambert_to_wgs(row):
    """
    Converts the x and y coordinates of this row into a Series of the
    longitude and latitude.
    
    """
    return pd.Series(pyproj.transform(lambert, wgs84, row['X'], row['Y']))
```

```python
transform_lambert_to_wgs(survey_data.loc[0])
```

```python
survey_data.apply(transform_lambert_to_wgs, axis=1)
```

```python
survey_data[["decimalLongitude" ,"decimalLatitude"]] = survey_data.apply(transform_lambert_to_wgs, axis=1)
```

```python
survey_data.head()
```

```python
survey_data.to_csv("Vermeersch_completed.csv", index=False)
```
